```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('m2.csv')

# Define custom color palette
colors = {'YouTube': 'red', 'Facebook': 'blue', 'Instagram': 'pink', 'WhatsApp': 'green', 'TikTok': 'black'}

# Create a count plot with custom colors
sns.countplot(x='Platform', data=data, palette=colors)
plt.title('Popularitas Platform Media Sosial')
plt.xlabel('Platform')
plt.ylabel('Jumlah Pengguna')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('m2.csv')

# Define custom color palette (optional)
# The error was due to TikTok having an empty string as its color.
# Changed to a valid color 'black' or any other desired color
colors = {'YouTube': 'red', 'Facebook': 'blue', 'Instagram': 'pink', 'WhatsApp': 'green', 'TikTok': 'black'}

# Get the platform counts and unique platform labels
platform_counts = data['Platform'].value_counts()
platform_labels = data['Platform'].unique()

# Map colors to platform labels
pie_colors = [colors[label] for label in platform_labels] # Create a list of colors corresponding to each platform label

# Create a pie chart with custom colors
plt.pie(platform_counts, labels=platform_labels, autopct="%1.1f%%", startangle=140, colors=pie_colors) # Use the mapped colors list here

# Customize the pie chart (optional)
plt.title('Popularitas Platform Media Sosial')
plt.axis('equal')  # Set aspect ratio for a circular chart

# Show the chart
plt.show()
```

```python
# Histogram untuk melihat distribusi usia
sns.histplot(data=data, x='Usia')
plt.title('Distribusi Usia Pengguna')
plt.xlabel('Usia')
plt.ylabel('Jumlah Pengguna')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('m2.csv')

# Filter data untuk platform YouTube
df_youtube = df[df['Platform'] == 'YouTube']

# Membuat histogram dengan warna merah
sns.histplot(data=df_youtube, x='Usia', bins=20, kde=True, color='red')
plt.title('Distribusi Pengguna YouTube Berdasarkan Usia')
plt.xlabel('Rentang Usia')
plt.ylabel('Jumlah Pengguna')
plt.show()
```
